```python
def flip_coin():
        from random import random
        return 'H' if random()< 0.5 else'T'
```

```python
flip_coin()
```

```python
def perform_experiment(experiments,n):
   results = []
   for x in range(n):
      result = experiments()
      results.append(result)
   return results
```

```python
perform_experiment(flip_coin,n=20)
```

```python
def count_matching(condition,results):
   matched=0
   for result in results:
     if condition(result):
       matched+=1
     return matched
```
